```python
## SPECIFYING COLUMNS DATA TYPES

data1=('1,2,3,4\n'
      '5,6,7,8\n'
      '9,10,11')
```

```python
print(data1)
```

```python
df=pd.read_csv(StringIO(data1),dtype=object) #If you use dtype=object, Pandas treats all columns as strings (str), even if they contain numbers.
```

```python
df
```

```python
df=pd.read_csv(StringIO(data1),dtype={'b':int,'c':float,'a':'Int64'})
df
```
